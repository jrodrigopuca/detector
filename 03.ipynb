### Capturar frames de YouTube

```python
import pafy
import matplotlib.pyplot as plt
import cv2
```

```python
url = 'https://www.youtube.com/watch?v=-ve-BbLnGno'
video = pafy.new(url)

print('[Title]',video.title)
best = video.getbest(preftype="mp4")
print('[URL]',best.url)
```

```python
miv=cv2.VideoCapture(best.url)
success, frame = miv.read()
i=0
while success:
    i+=1
    if (i==200):
        plt.imshow(frame)
    success, frame = miv.read()    
miv.release()
```
