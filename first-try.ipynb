```python
import numpy as np
import cv2
import datetime
import matplotlib.pyplot as plt
```

```python
# Seleccionar imagen
file="images/001.jpg"
imagen= cv2.imread(file)
rgb= cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
# Empiezo a capturar el tiempo
myTime =  datetime.datetime.now()

# Utilizo HOG
hog = cv2.HOGDescriptor()
hog.setSVMDetector( cv2.HOGDescriptor_getDefaultPeopleDetector() )

# Capturo los puntos en donde posiblemente se encuentre la persona
rects, weights = hog.detectMultiScale(rgb,winStride=(4, 4),padding=(8, 8),scale=1.05)

# Dibujo rectángulo
for (x, y, w, h) in rects:
    cv.rectangle(rgb,(x,y),(x+w,y+h),(0,255,0),10)    

# Mostrar
plt.imshow(rgb)

print("[INFO] puntos", rects)
print("[INFO] pesos", weights)
print("[INFO] tiempo: {}s".format((datetime.datetime.now() - myTime).total_seconds()))
```
